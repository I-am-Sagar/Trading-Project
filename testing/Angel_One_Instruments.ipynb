```python
import pandas as pd
import numpy as np
import requests
import json
from datetime import datetime
```

```python
# Get all instruments data which are available on Angel One

url = 'https://margincalculator.angelbroking.com/OpenAPI_File/files/OpenAPIScripMaster.json'
data = requests.get(url).json()
df = pd.DataFrame.from_dict(data)
df['expiry'] = pd.to_datetime(df['expiry'])
df = df.astype({'strike':float})
display(df)
```

```python
# We are interested in Equities and Mutual Funds as of now. So let's extract them. 

_, nse_equities_df = [x for i, x in nse_df.groupby(df["symbol"].str.contains('EQ'))]
display(nse_equities_df)

_, nse_mf_df = [x for i, x in nse_df.groupby(df["symbol"].str.contains('MF'))]
display(nse_mf_df)
```

```python
# Create two look up JSONs for equities for future use: Symbol_token and Token_symbol

symbol_token_dict = {}
token_symbol_dict = {}

for row in nse_equities_df.itertuples(index = True, name ='Pandas'):
    symbol = getattr(row, "symbol")
    token = getattr(row, "token")
    symbol_token_dict[symbol] = token
    token_symbol_dict[token] = symbol
    
# print(symbol_token_dict)
# print(token_symbol_dict)

with open('/Users/iamsagar/Desktop/Trading Project/data/instruments/eq_symbol_token.json', 'w') as fp:
    json.dump(symbol_token_dict, fp, ensure_ascii=False, indent=4)
    
with open('/Users/iamsagar/Desktop/Trading Project/data/instruments/eq_token_symbol.json', 'w') as fp:
    json.dump(token_symbol_dict, fp, ensure_ascii=False, indent=4)
```
